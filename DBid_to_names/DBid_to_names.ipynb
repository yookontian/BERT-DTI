```python
with open('/home/tian/Projects/BERT-DTI/KD-DTI-share/all_data_for_submit/DrugBank.test.json', 'r') as f:
    data = json.load(f)
```

```python
for instance in data:
    # print(data[instance])
    for triple in data[instance]['triples']:
        # print(triple)
        for drugs_and_targets in triple:
            if drugs_and_targets not in ['drug', 'target']:
                continue
            else:
                if not type(triple[drugs_and_targets]) is str:
                    raise TypeError(data[instance])
                if "####" in triple[drugs_and_targets]:
                    id = triple[drugs_and_targets].split("####")[1]
                    if drugs_and_targets == 'drug':
                        triple[drugs_and_targets] = DrugBank[id]
                    elif id in BioEntity_dict:
                        triple[drugs_and_targets] = BioEntity_dict[id]
```

```python
with open('DrugBank.test.json', 'w') as f:
    json.dump(data, f, indent=2, separators=(',', ': '))
```
